#  program to pop an element from the list and add to another list

```python
tasks = ["turn on computer", "insert password", "open program"]
```

```python
status = input("Have you done the task? (yes/no) = ")
```

```python
accomplishedTasks = []
```

```python
if status == "yes" and len(tasks)!=0:
    taskdone = tasks.pop()
    accomplishedTasks.append(taskdone)
elif status == "no" and len(tasks)!=0:                    #program to pop an element from the list and add in another
    print("complete the tasks first")
else:
    print("There are no task in the list")
tasks
```

```python
accomplishedTasks
```

# for loop

```python
for x in range(10):       # x start from 0 to 9
    print(x, "Aquib", end="\t")
```

```python
x           # shows the last value it holds i.e '9' in this case
```

```python
for x in range(1,10):       # x start from 1 to 9
    print(x, "Aquib")
```

```python
cars = ["audi","bentley","mehran"]
for car in cars:
    print("The car selected is", car.title()+".")
```

```python
for i in range(1,10,2):
    print(i)
```

```python
for i in range(9,1,-2):
    print(i)
```

```python
for w in "a","c","d","f":
    print(w)
```

```python
squareList= []
for value in range(1,11):
    square = value**2
    squareList.append(square)
print(squareList)
print(sum(squareList))
print(max(squareList))
print(min(squareList))
```

```python
for char in "Karachi":
    print(char)
```

```python
for char in "Karachi", "Pakistan":
    print(char)
```

```python
[x**2 for x in range(1,11)]     #list comprehension 
```

```python
[x**2 for x in range(1,11) if x % 2 ==0]        #not even no. printing
```

```python
y = [1,2,3,4,5,6]
x = int(input("Enter the number = "))

[i**2 for i in y if x==i]
```
