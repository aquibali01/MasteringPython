# nested conditions

```python
cnic = input("Do you have CNIC? (y/n) ")
voter_list = input("name in voter list? (y/n) ")
if cnic.lower() == "y" and voter_list.lower() == "y":
    print("You are allowed to vote")
else:
    print("Grow up first")
```

```python
day = input("Is this Sunday? (yes/no) ")
ticket = input("Do you have a ticket ? (yes/no) ")
permission = input("Do you have permission? (yes/no) ")
if day.lower() == "yes" and ticket.lower() == "yes" and permission.lower() == "yes":
    print("You are allowed to watch the match")
else:
    print("You are not allowed to watch")
```
