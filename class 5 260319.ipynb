```python
list = [1,3,5,8,10]                                #break at a certain point
for i in list:
    print(i)
    if i == 5:
        break
```

```python
list = [1,3,5,8,10]                                
for i in list:
    if i == 5:
        print(i)
        break
```

```python
list = [1,3,5,8,10]                                
for i in list:
    if i == 5:
        break
    print(i)
```

```python
list = [1,3,5,8,10]                   # 'continue ' skips the value in the condition                               
for i in list:
    if i == 5:
        continue
    print(i)
```

# for loop nested

```python
first_names = ["BlueRay ", "Upchuck ", "Lojack ",
"Gizmo ", "Do-Rag "]
```

```python
last_names = ["Zzz", "Burp", "Dogbone", "Droop"]
```

```python
full_names = []
```

```python
for f in first_names:
    for l in last_names:
        full_names.append(f+" "+l)
        
```

```python
for i in range(1,11):
    print("2*" + str(i) + "=" + str(i*2))
```

```python
for x in range(2,11):
    for y in range(1,11):
        print(str(x) + " * " + str(y) + " = " + str(x*y))
    print("===========",)
    
```

```python
start = int(input("Enter the start of tables : "))
end = int(input("Enter the end of tables : "))


for x in range(start,end):
    for y in range(1,11):
        print(str(x) + " * " + str(y) + " = " + str(x*y), end= " ")
    print("\n ")
```
