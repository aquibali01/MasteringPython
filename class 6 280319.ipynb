```python
def add():
    a = 12
    b = 13
    c = a + b
    print("result of c is : " + str(c))
```

```python
add()
```

```python
add()
print(a)         #it is not defined outside 
```

```python
def add():
    global a                      # it is defined globally
    a = 12
    b = 13
    c = a + b
    print("result of c is : " + str(c))
add()
print(a)
```

```python
d = 3000
def add():    # parameterless function
    global a                      
    a = 12
    b = 13
    c = a + b
    d = 100
    print(d)
    print("result of c is : " + str(c))
add()
print(a)
print(d)
```

```python
def addition(a,b):    # a & b are parameterised function
    c = a + b
    print(c)
addition(50,30)      # 50 & 30 are arguments
```

```python
def substraction(a,b):
    c = a - b
    print(c)
substraction(30,10)
```

```python
def abbaBeta(abba, beta):
    print("Abba is : " + abba + ", beta is :" + beta)
abbaBeta("nawaz", "arsalan")    # positional argument (postion jo define hain wohi rahegi)
```

```python
abbaBeta(beta="Arsalan",abba= "nawaz")    # keyword argument (postion jo jiski hai woh he hai)
```
